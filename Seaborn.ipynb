```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cs_df=pd.read_csv('C:\\Users\\jaiad\OneDrive\\Desktop\\ComputerSales.csv')
#print(sns.get_dataset_names())
crash_df=sns.load_dataset('car_crashes')
crash_df
```

```python
#Distribution Plot
sns.distplot(crash_df['not_distracted'], kde=False, bins=25)
```

```python
#Joint Plot
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

```python
#KDE Plot
sns.kdeplot(crash_df['alcohol'])
```

```python
#Pair plots
tips_df=sns.load_dataset('tips')
tips_df
sns.pairplot(tips_df, hue='sex',palette='Blues')
```

```python
#Rug Plot
sns.rugplot(tips_df['tip'])
```

```python
#Categorical Plots
#Bar Plots
sns.barplot(x='sex',y='total_bill', data=tips_df, estimator=np.median)#np.std,var,cov

```

```python
#Count plot
sns.countplot(x='sex', data=tips_df)
```

```python
#Box Plot
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.legend(loc=0)
#It is useful for finding the outlier detection
```

```python
#Violin plot
sns.violinplot(x='day', y='total_bill',data=tips_df, hue='sex',split=True)
```

```python
#Strip plot
plt.figure(figsize=(8,5))
sns.stripplot(x='day', y='total_bill', data=tips_df,jitter=True,hue='sex',dodge=True)
```

```python
#Swarm Plot

sns.swarmplot(x='day', y='total_bill',data=tips_df)
```

```python
#Palettes:
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex',palette='afmhot')
plt.legend(loc=1)
```

```python
#Matrrix Plots
#Heatmaps
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
crash_mx=crash_df.corr()
sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
flights=sns.load_dataset('flights')
flights=flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(flights, cmap='Blues')
```

```python
#Cluster map
iris=sns.load_dataset('iris')
species=iris.pop('species')
sns.clustermap(iris)
```

```python
#Pair Grid
iris=sns.load_dataset('iris')
iris_g=sns.PairGrid(iris,hue="species")
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
#Facet Grid
tips_fg=sns.FacetGrid(tips_df, col='time', row='smoker',height=4,aspect=1.3)
tips_fg.map(plt.hist, 'total_bill', bins=8)
```
