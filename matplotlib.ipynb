```python
#Pie Charts
import matplotlib.pyplot as plt
import random

fig_6 = plt.figure(figsize=(8, 5), dpi=100)
axes_6 = fig_6.add_axes([0.1, 0.1, 0.8, 0.8])  


types = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
         'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon', 'Fairy',
         'Ghost', 'Ice']

poke_num = [133, 109, 101, 98, 85, 77, 68, 66, 65, 60, 57, 54, 53, 51, 50, 50, 46, 45]

colors = []
for i in range(len(types)):
    rgb = (random.uniform(0, 0.5), random.uniform(0, 0.5), random.uniform(0, 0.5))
    colors.append(rgb)

explode = [0] * len(types)
explode[0] = 0.2 

wedges, texts, autotexts = plt.pie(poke_num, explode=explode, labels=types,
                                   colors=colors, autopct='%1.1f%%', shadow=True,
                                   startangle=140, textprops=dict(color="w"))

plt.legend(wedges, types, loc='center left', bbox_to_anchor=(1, 0.5))


```

```python

goog_data['Open'] = pd.Series([round(val, 2) for val in goog_data['Open']], index=goog_data.index)
goog_data['High'] = pd.Series([round(val, 2) for val in goog_data['High']], index=goog_data.index)
goog_data['Low'] = pd.Series([round(val, 2) for val in goog_data['Low']], index=goog_data.index)
goog_data['Close'] = pd.Series([round(val, 2) for val in goog_data['Close']], index=goog_data.index)
goog_data['Adj Close'] = pd.Series([round(val, 2) for val in goog_data['Adj Close']], index=goog_data.index)


stk_data = goog_data[-5:]


col_head = ('Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume')


stk_data_np = stk_data.to_numpy()


plt.figure(linewidth=2, tight_layout={'pad': .5}, figsize=(5, 3))
axes_8 = plt.gca()
axes_8.get_xaxis().set_visible(False)
axes_8.get_yaxis().set_visible(False)
plt.box(on=None)

ccolors = plt.cm.Blues(np.full(len(col_head), 0.2))

the_table = plt.table(cellText=stk_data_np, loc='center', colLabels=col_head, colColours=ccolors)
the_table.set_fontsize(14)
the_table.scale(3, 2.5)


plt.show()
```

```python
#Scatter Plot
np.random.seed(0)
x1 = np.random.randn(100)
y1 = np.random.randn(100)
sizes1 = 100 * np.random.rand(100)
colors1 = np.random.rand(100)

x2 = np.random.randn(100) + 5
y2 = np.random.randn(100) + 5
sizes2 = 100 * np.random.rand(100)
colors2 = np.random.rand(100)


fig, ax = plt.subplots(figsize=(10, 6))


scatter1 = ax.scatter(x1, y1, s=sizes1, c=colors1, alpha=0.6, edgecolors='w', cmap='viridis', label='Group 1')


scatter2 = ax.scatter(x2, y2, s=sizes2, c=colors2, alpha=0.6, edgecolors='w', cmap='plasma', label='Group 2')


cbar1 = plt.colorbar(scatter1, ax=ax)
cbar1.set_label('Color intensity Group 1')
cbar2 = plt.colorbar(scatter2, ax=ax)
cbar2.set_label('Color intensity Group 2')

ax.set_xlabel('X-axis label')
ax.set_ylabel('Y-axis label')
ax.set_title('Complex Scatter Plot Example')


ax.grid(True)


ax.legend()


```

```python
#3D surface 

from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')


surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')


cbar = fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
cbar.set_label('Intensity')


ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot Example')


plt.show()
```

```python
import mplfinance as mpf

dates = pd.date_range('2023-01-01', periods=100)
np.random.seed(42)
data = np.random.randn(100, 4).cumsum(axis=0) + 100
data = np.clip(data, 0, None) 

ohlc_data = pd.DataFrame(data, columns=['Open', 'High', 'Low', 'Close'], index=dates)
ohlc_data['Volume'] = np.random.randint(100, 1000, size=(100,))

ohlc_data['High'] = ohlc_data[['Open', 'Close']].max(axis=1) + np.random.rand(100)
ohlc_data['Low'] = ohlc_data[['Open', 'Close']].min(axis=1) - np.random.rand(100)


mpf.plot(ohlc_data, type='candle', volume=True, title='Candlestick Chart Example', style='charles')
```

```python
#HeatMap
import seaborn as sns



np.random.seed(42)
data = {
    'Age': np.random.randint(20, 80, 100),
    'BMI': np.random.uniform(18, 35, 100),
    'BloodPressure': np.random.uniform(80, 180, 100),
    'Cholesterol': np.random.uniform(150, 300, 100)
}

health_df = pd.DataFrame(data)

correlation_matrix = health_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Health Metrics Correlation')
plt.show()

```
